```python
import glob
```

```python
path_to_images = '/mnt/efs/data/'

list_of_BMW_image_filepaths = glob.glob(path_to_images+'**/*.jpg', recursive=True)
```

```python
len(list_of_BMW_image_filepaths)
```

```python
label = []
number_images = []
number_all_imgs=[]
for model_path in  glob.glob(path_to_images+'*/'):
    label.append(model_path.split('/')[-2])
    print(model_path.split('/')[-2])
    number=len(glob.glob(model_path+'/*/*/*/*/*.jpg'))
    number_images.append(number)
    print(number)
    number_all_imgs.append(number)
    
```

```python
sum(number_all_imgs)
label_sorted = sorted(zip(label,number_images))

label = []
number_images = []
for tupel in label_sorted:
    label.append(tupel[0])
    number_images.append(tupel[1])
print(label_sorted)
print(label)
print(number_images)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
figure = plt.figure()
sns.barplot(x=number_images,y=label)
plt.xticks(rotation='vertical')
figure.set_size_inches(20,12)
plt.show()

#TODO
# cool color style
# color by size
# display nunmbers on bars
```
